# Extract indentation raw data

Este programa tiene la finalidad de analizar todos las simulaciones de indentación y unirlos en un solo 
archivo de datos csv. Incluyendo la siguiente información para cada paso de tiempo de la simulación.

- Timestep
- Displacement
- Strain
- Force
- Stress
- Contact area
- Coordination analysis

Importamos librerías

```python
# Importación de librerías
from ovito.io import *
from ovito.modifiers import *
from ovito.pipeline import *
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from ClassIndentationVonMises import Indentation
```

```python
""" Definimos una función que genere los nombres de los archivos dump.indent e indent.dat en donde están las simulaciones """

# type_particle: 'hnp' for hollow nanoparticles or
#                'snp' for solid nanoparticles

def metadatos(orientation, diameter, silicon_percentage, type_particle):

    dumpfile = f"/home/mejia/Documents/Servicio_Social_2022_2023/hnp_si/simulaciones/{type_particle}/{int(diameter/10)}nm/{orientation}/{silicon_percentage}si-{100-silicon_percentage}ge/dump.indent"

    indentfile = f"/home/mejia/Documents/Servicio_Social_2022_2023/hnp_si/simulaciones/{type_particle}/{int(diameter/10)}nm/{orientation}/{silicon_percentage}si-{100-silicon_percentage}ge/indent.dat"


    metadata = { 'dumpfile': dumpfile, 
                 'indentfile': indentfile
                }
    
    return metadata
```

```python
""" Parámetros """

diameter = 100 # En Amstrong

"""Extraccción de datos"""

df = pd.DataFrame(columns=
        ['orientation', 'si-percentage', 'timestep', 'initdiam',
         'xeta', 'force', 'pote', 'displacement', 'strain',
         'contact-elliptical-area', 'elliptical-stress', 'percent-cn-1',
         'percent-cn-2', 'percent-cn-3', 'percent-cn-4',
         'percent-cn-5', 'percent-cn-6', 'percent-cn-7'])

for i, orientation in enumerate([100, 110, 111, 112]):

    for si_percentage in range(0,110,20):

        metadata = metadatos(orientation, 1, diameter, si_percentage, 'si-ge-coreshell')
        indentation = Indentation(metadata['dumpfile'], metadata['indentfile'])
        indentation.dataIndent['si-percentage'] = si_percentage
        indentation.dataIndent = indentation.dataIndent[['si-percentage'] + [ col for col in indentation.dataIndent.columns if col!='si-percentage']]
        indentation.dataIndent['orientation'] = orientation
        indentation.dataIndent = indentation.dataIndent[ ['orientation'] + [ col for col in indentation.dataIndent.columns if col != 'orientation']]
        indentation.setDisplacement()
        indentation.setStrain()
        indentation.setEllipticalStress()
        indentation.setEllipticalContactArea()
        indentation.setCoordinationAnalysis(2.8);
        indentation.dataIndent['elliptical-stress'].fillna(0, inplace=True)
        df = pd.concat([df, indentation.dataIndent], ignore_index = True)
        df.reset_index()
```

```python
filename = '/home/mejia/Documents/Servicio_Social_2022_2023/hnp_si/data_analysis/data/coreshell-data.csv'
df.to_csv(filename, index=False)
```
